# Apriori Algorithm:-

```python
#import the library for association Rule:-
#!pip install mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemset=apriori(df1,min_support=0.3,use_colnames=True)
frequent_itemset["length"]=frequent_itemset["itemsets"].apply(lambda x:len(x))
frequent_itemset
```

# Conclusion

# if we take min_support is 30% and confidance is 100% them the below result are shown

```python
rule=association_rules(frequent_itemset,min_threshold=1)
rule.head()
```

```python
frequent_itemset=apriori(df1,min_support=0.4,use_colnames=True)
frequent_itemset["length"]=frequent_itemset["itemsets"].apply(lambda x:len(x))
frequent_itemset
```

```python
rule=association_rules(frequent_itemset,min_threshold=0.7)
rule.head()
```
