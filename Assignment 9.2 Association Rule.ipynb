# Apply apriori Algorithm

```python
frequent_itemset=apriori(df,min_support=0.07,use_colnames=True)
frequent_itemset["length"]=frequent_itemset["itemsets"].apply(lambda x: len(x))
frequent_itemset
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rule=association_rules(frequent_itemset,min_threshold=0.9)
```

# Conclusion:-1 with min_support 0.07 and confidance 0.9 

```python
rule.head()
```
